```python
import pandas as pd 

def dataFrameCreator (studentsData, colLabel):
    """This functions collect the data and organize them in a table"""
    result = pd.DataFrame()
    result[colLabel[0]] = studentsData[0]
    result[colLabel[1]] = studentsData[1]
    result[colLabel[2]] = studentsData[2]
    result[colLabel[3]] = studentsData[3]
    result[colLabel[4]] = studentsData[4]
    result[colLabel[5]] = studentsData[5]
    result[colLabel[6]] = studentsData[6]
    result[colLabel[7]] = studentsData[7]
    
    fileSaver(result)  #This is a call to a function that saves the table as Excel Document.
    return result

def fileSaver (table):
    """Stores the file as Excel Document."""
    table.to_excel('c:/FileStored/studentRecord.xlsx', index = False)
    
def gradeAssignment(totalScores):
    """This function handles the assignment of Grade to student.
    The Function contains the letter grade for possible score
    obtainable by the student."""
    
    gradeList = ['A','B','C','D','E','F']
    for score in totalScores:
        if score >= 80:
            grade = gradeList[0]
        elif score >= 70:
            grade = gradeList[1]
        elif score >= 60:
            grade = gradeList[2]
        elif score >= 50:
            grade = gradeList[3]
        elif score >= 40:
            grade = gradeList[4]
        else:
            grade = gradeList[5]
    return grade
```

```python
#Sets of Global Variables
students = []
CA1, CA2, CA3, examList = [], [], [], []
totalCAScores, totalScores, grade = [], [], []

def tableCreator():
    """This function stores the student data in a list before making a call
    to the function that organize the data in table"""
    studentsRecords = [students,CA1, CA2, CA3, totalCAScores, examList, totalScores, grade]
    colNames = ['Names', 'CA 1', 'CA 2', 'CA 3', 'Total CA', 'Exam Score','Total','Grade'] 
    #Call to Data Frame Creator function
    finalResult = dataFrameCreator(studentsRecords, colNames)
    return finalResult
    
def finalComputation(test1, test2, test3, examScore):
    """This function states the student's tests and exams score and compute the 
    total. it also assigned grade to the score."""
    for score1, score2, score3, exam in zip (test1, test2, test3, examScore):
        #Computing the total test score
        totalTestScore = (score1+score2+score3)
        totalCAScores.append(totalTestScore)
        
        #Computing the Total Scores
        totalScore = totalTestScore + exam
        totalScores.append(totalScore)
        
        #Assigning the Grade
        studentsGrade = gradeAssignment(totalScores)
        grade.append(studentsGrade)
    
    
    #Call to the SubRoutine Function
    finalResult = tableCreator() 
    return finalResult

#creating a customize exception handling
class codeException(Exception):
    pass

def subProgram():
    try:
        studentName = input('Student Name: ')
        test1 = float(input('CA 1 Score: '))
        test2 = float(input('CA 2 Score: '))
        test3 = float(input('CA 3 Score: '))
        if (test1 > 10) or (test2 > 10) or (test3 > 10):
            raise codeException('Continuous Assessment Score can\'t be greater than 10')
        exam = float(input('Exam Score: '))
        if exam > 70:
            raise codeExceptionException('Examination Score can\'t be greater than 70')
    except ValueError as e:
        print(f'Invalid Input {e}')
    except Exception as e:
        print(f'Error! {e}')
    else:
        students.append(studentName)
        CA1.append(test1)
        CA2.append(test2)
        CA3.append(test3)
        examList.append(exam)
        
def starter ():
    print ('-'*60)
    print ("JOEMI TECHNOLOGIES RESULT COMPUTATION APP.")
    print ('Result Computation Made Easy....')
    print ('-'*60)
    start = True
    while start:
        commandToStart = input('Type Yes to Continue or No to Exit: ').lower()
        if commandToStart == 'no':
            start = False
            break
        try:
            if (commandToStart != 'yes') and (commandToStart != 'no'):
                raise Exception
        except:
            print('Invalid Input! Try again')
            continue
        else:
            subProgram()

        
def mainProgram():
    """Main body of the program that accepts input from user."""
    starter()
    finalResult = finalComputation(CA1, CA2, CA3, examList)
    return finalResult 
```

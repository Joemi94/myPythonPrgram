```python
#Program to simulate a multiple choice questions Bank
#Written by John Emeka Friday
#Cohort_16DA
#Created: 06 August, 2022.
#Modified: 27 August, 2022.

questionBank = {}   #Dictionary to store the multiple choice questions

def questionFunc():     
    #This function contains the questions bank of the program where the questions are evaluated and score"""
    score = 0
    questionList = ["Q1.What is the Capital of France?" \
"\n(a)Abuja" \
"\n(b)Paris" \
"\n(c)Lagos" \
"\n(d)None.", "Q2.Which state has only one neigbour?" \
"\n(a)Naimey" \
"\n(b)Cotonuo" \
"\n(c)Maine" \
"\n(d)None.", "Q3. Abuja is the Capital of Nigeria?"\
"\n(a)Yes" \
"\n(b)No" \
"\n(c)I don't Know" \
"\n(d)is not my business."]
    answerList = ['b','c','a']
    for question, answer in zip (questionList, answerList):
        questionBank[question] = answer
    for key in questionBank:
        print(key)
        userOption = input("Type your answer(a/b/c/d):")
        if userOption == questionBank[key]:
            score += 1
    return score
```

```python

def starter ():
    #Introduction and Instruction
    print("_"*80)
    print("Welcome to Joemi Academy Aptitude Test")
    print("INSTRUCTION: Choose the Correct Option from the letter number A to D\n"\
    "Each questions Carries equal mark.")
    print("""
    1) Enter START to Start Quiz
    2) Enter QUIT to exit
    3) You only have 2 ATTEMPTS""")
    print("_"*80)

def subProgram():
    """
    This functions generate the questions by calling the question bank function.
    """
    counter = 2    #handles the 2 attempt provider for user
    
    quizScore = 0    #initializing score
    
    quizCommencement = True
    while quizCommencement:
        quizScore = questionFunc()     #place a call to the questions bank
        
        counter -=1
        if counter == 1:
            print(f"You have {counter} more attempt.Type YES to re-take Quiz or No to submit")
            commandToStart = input(": ").lower()
            if commandToStart == "no":
                quizCommencement = False
                print("Successfully Submitted!")
                break
        elif counter == 0:
            print("Successfully Submitted!")
            quizCommencement = False
            break

    print(f"You Score {quizScore} out of {len(questionBank)}")   
    return quizCommencement

def mainProgram ():
    #Main Body of the Program starts here
    starter ()   #place a call to the starter function
    quizCommencement = True
    while quizCommencement:
        print("Enter START to start Quiz or QUIT if you are not ready")
        commandToStart = input(": ").lower()
        if commandToStart == "quit":
            quizCommencement = False
            print("Try Re-taking the Quiz before Deadline")
            break
        try:
            if commandToStart != "start":
                raise Exception 
        except:
            print ("Invalid Input!")
            continue
        else:
            quizCommencement = subProgram() #place a call to the subProgram function
            break
```
